## base normalized

```python
def create_dataset(data, drop, label_name, test_size=0.2):
    X = data.drop(drop, axis=1)
    y = data[label_name]
        
    return train_test_split(X, y, test_size=test_size)
```

```python
X_train, X_test, y_train, y_test = create_dataset(base_normalized, ['id', 'time', 'day_id', 'mood', 'label'], 'label')
```

```python
def train_rf(xtrain, xtest, ytrain, ytest, max_depth=None, n_estimators=100, random_search=False, cross_validation=False, epochs=10):
    avg = 0
    
    if random_search: 
        
        rf = RandomForestClassifier()

        param_dist = {'n_estimators': randint(100,200),
          'max_depth': randint(5,15)}

        rand_search = RandomizedSearchCV(rf, 
                             param_distributions = param_dist, 
                             n_iter=50, 
                             cv=5)

        rand_search.fit(xtrain, ytrain)
        
        rf = rand_search.best_estimator_
        print('Best hyperparameters:',  rand_search.best_params_)
    
    else: 
         rf = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth)
    
    for e in range(epochs):
        if cross_validation:
            avg += np.mean(cross_val_score(rf, xtrain, ytrain, cv=5))
        else: 
            rf.fit(xtrain, ytrain)
            avg += rf.score(xtest, ytest)
    
    # returns averaged score and last model
    return avg / epochs, rf
```

```python
# no randomized search, no crossvalidation
train_rf(X_train, X_test, y_train, y_test)
```

```python
# with cross-validation
train_rf(X_train, X_test, y_train, y_test, cross_validation=True)
```

```python
# with randomized search + cross validation
train_rf(X_train, X_test, y_train, y_test, random_search=True)
```
