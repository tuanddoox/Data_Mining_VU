```python
# original data with features 
added_features = pd.read_csv('pre_processed_dataset/20230415_df_added_features.csv')
base_normalized = pd.read_csv('pre_processed_dataset/20230415_df_base_normalized.csv')
full_features = pd.read_csv('pre_processed_dataset/20230415_df_full_features.csv')
```

## without feature engineering

```python
def create_dataset(data, drop, label_name, test_size=0.2):
    X = data.drop(drop, axis=1)
    y = data[label_name]
        
    return train_test_split(X, y, test_size=test_size)
```

```python
X_train, X_test, y_train, y_test = create_dataset(base_normalized, ['id', 'time', 'day_id', 'mood', 'label'], 'label')
```

```python
def train_rf(xtrain, xtest, ytrain, ytest, epochs=10, max_depth=None, n_estimators=100, random_search=False):
    avg = 0
    
    if random_search: 
        
        rf = RandomForestClassifier()

        param_dist = {'n_estimators': randint(50,500),
          'max_depth': randint(1,20)}

        rand_search = RandomizedSearchCV(rf, 
                             param_distributions = param_dist, 
                             n_iter=5, 
                             cv=5)

        rand_search.fit(xtrain, ytrain)
        
        rf = rand_search.best_estimator_
        print('Best hyperparameters:',  rand_search.best_params_)
    
    else: 
        rf = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth)
    
    for e in range(epochs):

        # create new model and train
        rf.fit(xtrain, ytrain)
        
        # score model
        score = rf.score(xtest, ytest)
        
        # output score
        print('accuracy at epoch {}: {}'.format(e, score))
        
        avg += score
                
    return avg / epochs, rf
```

```python
train_rf(X_train, X_test, y_train, y_test)
```

```python
train_rf(X_train, X_test, y_train, y_test, epochs=100, random_search=True)
```

## ARIMA

```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import adfuller
```

```python
added_features = pd.read_csv('pre_processed_dataset/20230415_df_added_features.csv', index_col='day_id')
added_features = added_features.dropna()
```

```python
plot_acf(added_features.mood)
plt.show()
```

```python
f = plt.figure()
ax1 = f.add_subplot(121)
ax1.set_title('first order differencing')
ax1.plot(added_features.mood.diff())
```

```python
result = adfuller(added_features.mood)
print(result[1])

result = adfuller(added_features.mood.diff().dropna())
print(result[1])
```

```python
# 770/1155 train/test split
x_train = added_features.drop(['mood', 'id', 'time'], axis=1).iloc[:770]
x_test = added_features.drop(['mood', 'id', 'time'], axis=1).iloc[770:]
y_train = added_features.mood.iloc[:770]
y_true = added_features.mood.iloc[770:]

index = added_features.mood.iloc[770:].index
```

```python
arima = ARIMA(endog=y_train, exog=x_train, order=(1, 0, 2))
model = arima.fit()
```

```python
model.summary()
```

```python
plt.plot(model.predict())
```

```python
plt.plot(added_features.mood)
```

```python
y_pred = pd.Series(model.forecast(385, exog=x_test))
y_pred = pd.Series(y_pred.values, index=index)
y_pred
```

```python
y_pred, y_true
```

```python
mape = np.mean(np.abs(y_pred - y_true) / np.abs(y_true))
mae = np.mean(np.abs(y_pred - y_true))
mpe = np.mean((y_pred - y_true)/y_true)
rmse = np.mean((y_pred - y_true)**2)**.5
corr = np.corrcoef(y_pred, y_true)[0,1]
```
